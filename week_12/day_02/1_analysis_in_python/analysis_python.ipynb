```python
demo_data_missing = pd.read_csv("data/life_expectancy_and_income_missing.csv")

demo_data_missing
```

```python
# Summarise the missing values
demo_data_missing.isna().sum()
```

```python
# Can fill the NA's with 0
demo_data_missing.fillna(0)
```

```python
# Can pass in a dictionary for what keys for what variables

# Imputing the mean

demo_data_missing.fillna({
    'life_expectancy': demo_data_missing['life_expectancy'].mean(),
    'income_per_person': demo_data_missing['income_per_person'].mean()
})
```

```python
# Group by the country

(
    demo_data_missing.fillna({
        'life_expectancy': demo_data_missing
        .groupby('country')['life_expectancy']
        .transform('mean'),
        'income_per_person': demo_data_missing
        .groupby('country')['income_per_person']
        .transform('mean')
    })
)
```

```python
demo_data_missing.groupby('country')['life_expectancy'].transform('mean')
```

```python
# Drop NA
demo_data_missing.dropna()
```
