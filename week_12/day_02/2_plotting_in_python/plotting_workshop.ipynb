# Plotting Lovely Plots

```python
import pandas as pd
```

```python
demo_data = pd.read_csv("data/life_expectancy_and_income.csv")

demo_data.head()
```

```python
demo_data_2019 = demo_data.loc[demo_data.year == 2019].copy()

demo_data_2019.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [5, 3]
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.size'] = 10
```

```python
plt.scatter(x = 'income_per_person', y = 'life_expectancy', data = demo_data_2019)
```

```python
plt.scatter(x = 'income_per_person', y = 'life_expectancy', data = demo_data_2019)
plt.xlabel('Income per Person (£)')
plt.ylabel('Life Expectancy')
plt.title('Income per Person vs Life Exp')
```

```python
demo_data_2019.loc[:, 'income_per_person_1000'] = demo_data['income_per_person']/1000

plt.scatter(x = 'income_per_person_1000', y = 'life_expectancy', data = demo_data_2019)
plt.xlabel('Income per Person (£1000)')
plt.ylabel('Life Expectancy')
plt.title('Income per Person vs Life Exp')
plt.xlim(0,10)
plt.yticks(50, 55, 60, 65, 70, 75, 80, 85)
```

```python
plt.hist(x = 'income_per_person_1000', data = demo_data_2019)
```

```python
plt.boxplot(x = 'income_per_person_1000', data = demo_data_2019)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x = 'income_per_person_1000', y = 'life_expectancy', data = demo_data_2019)
```

```python
sns.histplot(x = 'income_per_person', data = demo_data_2019)
```

```python
sns.boxplot(x = 'income_per_person', data = demo_data_2019)
```

```python
demo_data_japan = demo_data.loc[demo_data['country'] == 'Japan']
```

```python
demo_data_japan.head()
```

```python
sns.lineplot(x = 'year', y = 'fertility_rate', data = demo_data_japan)
plt.title('Fertility Rate over Time, Japan')
```

```python
sns.barplot(x = 'country', y = 'life_expectancy', data = demo_data_2019.head(5))
```

```python
ordered_data = demo_data_2019.head(5).sort_values('life_expectancy')['country']

sns.barplot(x = 'country', y = 'life_expectancy',
            data = ordered_data_2019.head(5),
            order = ordered_data);
```

```python
last_20 = demo_data.loc[demo_data['year'] > 2000].copy()
```

```python
last_20
```

```python
import numpy as np
```

```python
sns.barplot(y = 'year', x = 'income_per_person', data = last_20,
            estimator = np.mean,
            ci = False,
            orient = 'h').invert_yaxis();
plt.title("Average income per person by year from 2000 to 2019")
plt.xticks(rotation = 45);
```
