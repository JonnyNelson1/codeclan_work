# Question 17

```python
# Obtain a table showing all employees details,
# together with a column team_name containing the name of their team.

(
    employees
    .merge(teams
    .loc[:, ['id', 'name']],
           how = 'left',
           left_on= 'team_id',
           right_on = 'id')
    .drop(columns = ['team_id', 'id_y'])
    .rename(columns={
        "id_x" : "id",
        "name" : "team_name" 
        })
)
```

# Question 18

```python
# Obtain a table showing team_name,
# together with a count of the num_employees in each team.

(
    employees
    .merge(teams
    .loc[:, ['id', 'name']],
           how = 'left',
           left_on= 'team_id',
           right_on = 'id')
    .drop(columns = ['team_id', 'id_y'])
    .rename(columns={
        "id_x" : "id",
        "name" : "team_name" 
        })
    .groupby('team_name')
    .id.count()
    .reset_index()
    .rename(columns =  {'id' : 'count'})
)
```

# Question 19

```python
# Obtain a table showing the id, first_name, last_name and department
# of any employees lacking both a local_account_no and local_sort_code,
# in their pay_details.

employees\
    .merge(
        pay_details.loc[pay_details[['local_account_no', 'local_sort_code']].isna().all(axis='columns')],
        how='inner',
        left_on='pay_detail_id',
        right_on='id'
    )\
    .rename(columns={'id_x': 'id'})\
    .loc[:, ['id', 'first_name', 'last_name', 'department']]
```
